```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
stdAge = (22,22,23,19,20,29,33,20,20,23,21,21,21,20,20,20)

df = pd.DataFrame({'Name': ('Faisal','Ahmed','Khalid','Faisal','Ahmed','Khalid'),'Age':(21,22,32,21,22,32),'GPA':(3.4,3.2,3.9,3.6,3.7,3.4)})

plt.scatter(
    df['Age'], df['GPA'],
    color='b', alpha=0.5, label='Student'
)

plt.title('Relationship Between Age and GPA', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.legend()

plt.show();




plt.hist(stdAge, bins=30, color='skyblue', edgecolor='white')

plt.title('Histogram of Age', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show();
```

## Q2: Create pie chart with equal explode values.

```python
mark = df['GPA'].value_counts().sort_values(ascending=True)

plt.pie(
    mark, labels=mark.index,
    autopct='%1.1f%%', startangle=90,
    explode = [0.3,0.3,0.3,0.3,0.3]
)

plt.title('Distribution of GPA\'s', fontsize=15)

plt.show();
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(
    df['Age'], df['GPA'],
    color='b', alpha=0.4, label='Student',s=50,
    marker='o'
)

plt.title('Relationship Between Age and GPA', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.legend()

plt.show();
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(
    df['Age'], df['GPA'],
    color='b', alpha=0.4, label='Student',s=50,
    marker='o'
)

plt.title('Relationship Between Age and GPA', fontsize=16,loc='left')
plt.xlabel('Age', fontsize=16)
plt.ylabel('GPA', fontsize=16)
plt.legend()

plt.show();
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(20, 10))

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0].plot(x, y1, color='r')
axs[0].set_title('Plot 1')

axs[1].plot(x, y2, color='b')
axs[1].set_title('Plot 2')

axs[2].plot(x, y1, color='g')
axs[2].set_title('Plot 3')



plt.show();
```
